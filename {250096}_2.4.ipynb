```python
import pandas as pd
import yfinance as yf
def SMA(s, w):
  return s.rolling(w).mean()
def EMA(s, w):
  return s.ewm(span = w, adjust = False).mean()
def MACD(s):
  first = 12
  second = 26
  signal = 9
  MACD_line = EMA(s,first) - EMA(s,second)
  Signal_line = EMA(MACD_line, signal)
  return MACD_line, Signal_line
def RSI(s,w):
  delta = s.diff()
  gain = delta.where(delta > 0, 0)
  loss = -delta.where(delta < 0, 0)
  avg_gain = gain.rolling(w).mean()
  avg_loss = loss.rolling(w).mean()
  rs = avg_gain / avg_loss
  rsi = 100 - (100 / (1 + rs))
  return rsi
def Bollinger(s,w = 20):
  middle = SMA(s, w)
  std = s.rolling(w).std()
  upper = middle + 2 * std
  lower = middle - 2 * std
  return middle, upper, lower
def Stochastic_Osc(high, low, close, w):
  lowest = low.rolling(w).min()
  highest = high.rolling(w).max()
  K = 100 * ((close - lowest)/(highest - lowest))
  D = SMA(K,3)
  return K,D

d = yf.download("RELIANCE.NS", period="2y", interval="1d")
close = d['Close']
high = d['High']
low = d['Low']
N = int(input("Enter the period : \n"))
ML, SL = MACD(close)
M,U,L = Bollinger(close)
K, D = Stochastic_Osc(high,low,close,N)
print("SMA is : \n", SMA(close,N).tail())
print("EMA is : \n", EMA(close,N).tail())
print("MACD Line is \n: ", ML.tail())
print("Signal Line is : \n", SL.tail())
print("RSI is : \n", RSI(close,N).tail())
print("The Bollinger Band is : \n", M.tail())
print(U.tail())
print(L.tail())
print("The Oscillator is : \n", K.tail())
print("SMA for 3 day periods of SOC : \n", D.tail())
```
